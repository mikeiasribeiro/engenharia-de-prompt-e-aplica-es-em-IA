calculadora simples

```
num = int(input("Digite um número: "))

print("\nSoma")
for i in range(1, 11):
    print(num, "+", i, "=", num + i)

print("\nSubtração")
for i in range(1, 11):
    print(num, "-", i, "=", num - i)

print("\nMultiplicação")
for i in range(1, 11):
    print(num, "*", i, "=", num * i)

print("\nDivisão")
for i in range(1, 11):
    print(num, "/", i, "=", num / i)
```
